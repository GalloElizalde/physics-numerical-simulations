# Bees Dynamics Analysis



## General informations

### Data description

This notebook proposes to analyze data describing the 2D motion of a system of two bees, in different situations. These data were generated by a numerical simulation with tunable parameters, such as an attraction force between the two bees, the presence of a field of flowers (attracting the bees too), or some random noise in the trajectory of each bees. More precisely, the following parameters will be considered:
 + `BOX`       : if enabled, there is a box (*i.e.* a limited area) in which the bees can live (bees are simply reflected when hiting a wall of the box)
 + `LBOX`      : length of the box, which is chosen to be a square
 + `ATTRACTION`: if enabled, bees attract each other (spring force)
 + `g_ATTRA`   : strenght of the mutual attraction
 + `FIELD`     : if enabled, there is a field of flowers organised on a regular lattice
 + `g_FIELD`   : strengh of individual flower attraction
 + `NOISE`     : if enabled, some random noise is added in individual bee trajectories
 

The data contains the following variables:
 + time (in ms) : `tms`
 + x-position of bee 1 and bee 2: `Rx1` and `Rx2`
 + y-position of bee 1 and bee 2: `Ry1` and `Ry2`
 + x- and y- velocities of bee 1 and bee 2 : `Vx1`, `Vx2`, `Vy1`, `Vy2`


The first section of the notebook proposes to study data that were already generated, while the second part focuses on data generation using an external module, in order to see how the model parameters affect the system.

## Analyzing existing data

Name of the dataframe: `beesMotion_BOX-1_LBOX-300_ATTRACTION-1_gATTRA-0.10_FIELD-0_gFIELD-2.80_NOISE-0_RUN-0.csv`.

```python
# Loading the fisrt dataframe and exploring it
path = "beesMotion_BOX-1_LBOX-300_ATTRACTION-1_gATTRA-0.10_FIELD-0_gFIELD-2.80_NOISE-0_RUN-0.csv"
df = pd.read_csv(path)
print(f"The dimension of the dataframe is: {df.shape}")

df.head()
```

```python
# Function to get parameters from the name of the csv

def param_values(csv):
    '''
    This function takes as argument
    the name of a csv 'csv' and 
    returns a dictionary 'dictionary' 
    storing the parameter simulation
    values as {param:val}
    '''
    # Define empty dictionary

    dictionary = {}

    
    # Remove the prefix "beesMotion_" and suffix ".csv"
    
    name = csv.replace("beesMotion_", "")
    name = name.replace(".csv", "")
    
    
    # List of param-val using split
    parameters = name.split("_")


    # Loop over param-val

    for param in parameters:
        pair = param.split("-")
        dictionary[pair[0]] = pair[1]


    return dictionary



# Dictionary of the cvs

dictionary = param_values(path)
print("This is the dictioary: ", dictionary)
```

```python
# Ploting the trajectory of Rx and Ry 

plt.title("Trajectory of Two Bees")
plt.plot(df.Rx1.values, df.Ry1.values, label = "Bee 1")
plt.plot(df.Rx2.values, df.Ry2.values, label = "Bee 2")
plt.ylabel("y")
plt.xlabel("x")
plt.legend()
plt.grid()


# Saving fig

plt.savefig("trajectory_two_bees.png", dpi = 300)
```

```python
# Adding a new column to the df containing distance between the two bees

del_x = df.Rx1 - df.Rx2
del_y = df.Ry1 - df.Ry2
df["Distance"] = np.sqrt(del_x ** 2  +  del_y ** 2)


# Plotting the distance between 2 bees

plt.title("Distance Between Two Bees")
plt.plot(df.tms, df.Distance, lw = 1.4)
plt.xlabel("time")
plt.ylabel("distance (ms)")
plt.grid()
plt.savefig("distance_two_bees.png", dpi = 300)
```

**1.5 [2pts]** We would like to understand how position and velocities are correlated between bee 1 and bee 2. Compute the velocity $v = \sqrt{v_x^2 + v_y^2}$ and the position $r = \sqrt{r_x^2 + r_y^2}$, for each bee. Produce a plot of $r_1$ versus $r_2$ and another plot $v_1$ versus $v_2$. What would you conclude about how correlated are bee 1 and 2?

```python
# Adding individual norm of velocity and position to the dataframe

# Position
df["norm_R1"] = np.sqrt( df.Rx1**2  +  df.Ry1**2 )
df["norm_R2"] = np.sqrt( df.Rx2**2  +  df.Ry2**2 ) 
# Velocity
df["norm_V1"] = np.sqrt( df.Vx1**2  +  df.Vy1**2 )
df["norm_V2"] = np.sqrt( df.Vx2**2  +  df.Vy2**2 )


# Plotting Correlations 

plt.figure(figsize = (14,5))

#PLot 1
plt.subplot(1,2,1)
plt.title("Trajectory Correlation")
plt.scatter(df.norm_R1, df.norm_R2, s = 0.5, c = df.tms, cmap = "magma")
plt.xlabel("$r_1(t)$")
plt.ylabel("$r_2(t)$")
plt.grid()

#Plot 2
plt.subplot(1,2,2)
plt.title("Velocity Correlation")
plt.scatter(df.norm_V1, df.norm_V2, s = 0.5, c = df.tms, cmap = "magma")
plt.xlabel("$v_1(t)$")
plt.ylabel("$v_2(t)$")
plt.colorbar()
plt.grid()

plt.tight_layout()
plt.savefig("correlation_two_bees.png", dpi = 300)
```

```python
# MAsk for finding the duration for which the bees are closer than 5

mask = df.Distance.values < 5 
time = df["tms"][mask]


# Total duration = number of steps * step value

del_time =  (time.iloc[-1] - time.iloc[0]) * .05
percentage =  (100/ df["tms"].size) * time.size 
print(f'The 2 bees are closed to each other during {del_time:.0f} ms, representing {percentage:0.1f} %')
```

## Simulating data using an external package

Importing the external package `bees_simulation.py` as `bs` and running a simulation using the function `run_bees_simulation`.

# Generate simulations (without noise) enabling a flowers field, for 3 different interaction strengh values `g_FIELD`, namely `0.0`, `2.8` and `5`.

```python
# Generating simulations (without noise) enabling a flowers field 

datas = []
g_fields = [0.0, 2.8, 5]  #3 different interaction strengh values `g_FIELD` = `0.0`, `2.8` and `5`.
for g in g_fields:  
    name = bs.run_bees_simulation(steps=10000, noise=0, field=1, g_field=g)
    d = pd.read_csv(name)
    datas.append(d)
```

```python
# Overlaying the bee 1 trajectory of each simulation

plt.title("Trajectories of Bee 1 for Different Flower Attraction")
for i, data in enumerate(datas):
    plt.plot(data.tms, data.Rx1, label = f"g_FIELD = {g_fields[i]}")
    
plt.ylabel("Rx1")
plt.xlabel("time (ms)")
plt.legend()
plt.grid()

plt.savefig("trajectory_bee_1_different_g_FIELD.png", dpi = 300)
```

**BONUS QUESTION 2.3 [2 pts]** Write a function which takes two datasets (*i.e.* two simulation setups) and plots as a function of time the difference between the 2 datasets of the distances between bees $r_{12} = \sqrt{(x_1-x_2)^2 + (y_1-y_2)^2}$;
  
Test it on the data sets with `g_FIELD=0` and `g_FIELD=5`.

```python
#

def plot_diff(data1, data2):
    '''
    Function which takes two datasets and plots 
    as a function of time the difference between 
    the 2 datasets of the distances between bees
    '''
    
    # Distance between bees
    plt.figure()
    r12_d1 = np.sqrt( (data1.Rx2-data1.Rx1)**2 + (data1.Ry2-data1.Ry1)**2)
    r12_d2 = np.sqrt( (data2.Rx2-data2.Rx1)**2 + (data2.Ry2-data2.Ry1)**2)
    plt.title("Distamce Between Bees")
    plt.plot(data1.tms, r12_d1 - r12_d2, label = "Distance")
    plt.xlabel("time (ms)")
    plt.ylabel("R2-R1")
    plt.grid()
    plt.legend()
    plt.savefig("distance_between_two_bees.png", dpi = 300)
    
    # Mean velocity
    plt.figure()
    v12_d1 = 0.5 * np.sqrt( (data1.Vx1+data1.Vx2)**2 + (data1.Vy1+data1.Vy1)**2)
    v12_d2 = 0.5 * np.sqrt( (data2.Vx1+data2.Vx2)**2 + (data2.Vy1+data2.Vy2)**2)
    plt.title("Mean Velocity")
    plt.plot(data1.tms, v12_d1 - v12_d2, label = "Mean")
    plt.xlabel("time (ms)")
    plt.ylabel("V2-V1")
    plt.grid()
    plt.legend()
    plt.savefig("mean_velocity_two_bees.png", dpi = 300)

```
